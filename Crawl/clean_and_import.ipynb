```python
import pandas as pd
import re
import os
from dotenv import load_dotenv
from label_studio_sdk import LabelStudio

load_dotenv()
```

```python
df = pd.read_json("ner2_raw.json")
```

```python
sentences_set = [re.split(r'\.(?=\s[A-Z])', sent) for sent in df['text']]
sentences = []
for item in sentences_set:
    [sentences.append(sent.strip()) for sent in item if sent != '']    
    
```

```python
tasks = []
for item in sentences:
    task = {
        "data": {
            "text" : item
        }
    }
    tasks.append(task)
    
```

```python
PROJECT_ID = 7
client = LabelStudio(base_url=os.getenv('LABEL_STUDIO_URL'), api_key=os.getenv('LABEL_STUDIO_API_KEY'))
resp = client.projects.import_tasks(
    id = PROJECT_ID,
    request=tasks,
    return_task_ids=True
)
print(resp)
```
